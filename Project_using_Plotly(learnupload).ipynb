```python
import pandas as pd

# Load files
latlong = pd.read_csv("district wise centroids.csv")
census = pd.read_csv("india-districts-census-2011.csv")

# Merge on District (left = latlong, right = census)
final_df = latlong.merge(
    census,
    left_on="District",
    right_on="District name",
    how="inner"
)

# Derived columns
final_df["sex_ratio"] = round((final_df["Female"] / final_df["Male"]) * 100, 2)
final_df["literacy_rate"] = round((final_df["Literate"] / final_df["Population"]) * 100, 2)

# Keep useful columns
columns_needed = [
    "State", "District", "Latitude", "Longitude",
    "Population", "Male", "Female", "Literate",
    "sex_ratio", "literacy_rate"
]
final_df = final_df[columns_needed]

# Save to CSV for Streamlit
final_df.to_csv("Final india.csv", index=False)

print("Final india.csv saved!")
```
